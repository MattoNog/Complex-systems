## **La ecuación de Burgers**





La ecuación de Burgers

\begin{equation}
    \frac{\partial u}{\partial t} + u \frac{\partial u}{\partial x} - \nu \frac{\partial^2 u}{\partial x^2} = 0.                          
\end{equation}

Forma parte de un grupo de ecuaciones no-lineales en derivadas parciales
"universales" que aparecen en muchos contextos.

En rigor, fue introducida en 1915 por Bateman [1], quien propuso una
solución de estado estacionario para el problema. Más
tarde, en 1948, el físico holandés J. M. Burgers, propuso
esta ecuación como modelo matemático de turbulencia [2,
3] y debido a su amplia contribución, la ecuación lleva su
nombre. A pesar de ser una versión simplificada de la ecuación de Navier Stokes,
se caracteriza por tener un término
dependiente del tiempo, un término de advección y un
término de difusión, con lo cual guarda la no-linealidad y las principales características de esta.
Se utiliza muchas veces como modelo de juguete para testear implementaciones numéricas de flujos turbulentos.

En su forma mas general (con el término de difusión) la ecuación es del tipo parabólica, mientras que en el caso ideal (invisido) es una ecuación hiperbólica. En ambos casos, resulta útil buscar leyes de conservación, de camino a las soluciones.

[1] Bateman, H., Some recent researches on the motion of
fluids, Monthy Wather Rev. 43, 163-170 (1915).

[2] Burgers, J. M., A mathematical model illustrating the
theory or turbulence, Advances in Applied Mechanics 1,
171-199 (1948).

[3] Burgers, J. M., Mathematical examples illustrating
relations occuring in the theory of turbulent fluid motion,
Trans. Roy. Neth. Acad. Sci. Amsterdam 1. 17, 1-53
(1939).

Un poco de desarrollo teórico:

$$\frac{\partial u}{\partial t} + u \frac{\partial u}{\partial x} - \nu \frac{\partial^2 u}{\partial x^2} = 0$$

Uso que $u \frac{∂u}{∂x} = \frac{∂}{∂x}(\frac{u^2}{2})$ debido a la regla de la cadena.

$$\frac{\partial u}{\partial t} = - \frac{\partial}{\partial x} (\frac{u^2}{2} - \nu u_x )$$

Integro en el contorno:

$$  \int_{0}^{L} ∂_t u \,dx  = - \int_{0}^{L} ∂_x (\frac{u^2}{2} - \nu u_x ) \,dx $$


$$ ∂_t \int_{0}^{L} u \,dx = - \int_{0}^{L} ∂_x (\frac{u^2}{2} - \nu u_x ) \,dx $$

Uso que $p= u ⋅ du$ y la periodicidad en el contorno (al integrar se va a anular):

$$ \frac{∂p}{∂t} = 0 $$

Considero $\nu = 0$ en Burgers

$$\frac{\partial u}{\partial t} + u \frac{\partial u}{\partial x} = 0$$

Multiplico por $u$

$$ u \frac{\partial u}{\partial t} + u^2 \frac{\partial u}{\partial x} = 0$$

Ahora lo puedo escribir de dos formas diferentes:

$$ \frac{∂ (\frac{u²}{2})}{\partial t} + u \frac{∂ (\frac{u²}{2})}{\partial x} = \frac{d}{dt}(\frac{u²}{2})$$

y

$$ \frac{∂ (\frac{u²}{2})}{\partial t} + \frac{∂ (\frac{u^3}{3})}{\partial t} = 0$$

De la primero saco la conservación local ya que $\frac{d}{dt}(\frac{u²}{2}) = 0$
y de la segunda se puede notar que con los correctos límites de integración se conserva la energía de forma global.

Primero busco la forma de gradientes de Burgers, derivo por x (uso que $\frac{∂u}{∂x} = ∇u$):

$$ \frac{∂ (\nabla u)}{∂t} = - (∇ u)^2 + ν \frac{∂^2(∇u)}{∂x^2}$$

En el caso de una viscosidad muy baja ($ν << 1$) el segundo término de la derecha decae antes que el primero. Por lo tanto al tener una divergencia a $-∞$. Es decir que $-(∇u)^2 = -∞$ osea que las diferencias locales de velocidades del fulído van a ser muy grandes. Estas diferencias se les llaman frentes de ondas.

Transformación de Hopf:

$u = -2 ν \frac{∂_x \phi}{ϕ} = -2 ν ∂_x ln(ϕ) = ∂_x(Φ) $

$$ \frac{\partial^2 Φ}{∂t∂x} + \frac{\partial Φ}{\partial x} ⋅ \frac{\partial^2 Φ}{\partial x^2}- ν \frac{\partial^3 Φ}{\partial x^3} = 0 $$

Usando la regla de la cadena me queda:

$$ \frac{\partial^2 Φ}{∂t∂x} + \frac{1}{2} ⋅ (\frac{\partial Φ}{\partial x})^2- ν \frac{\partial^3 Φ}{\partial x^3} = 0 $$

Uso que $Φ = -2 ν ln(ϕ)$ y lo meto en la anterior expresión. Haciendo la regla de la cadena pertinente llego a:

$$ -2\frac{\nu}{ϕ} \frac{∂ϕ}{∂t} + 2 \frac{\nu^2}{ϕ} \frac{∂^2ϕ}{∂x^2} = 0$$
$$ \frac{∂ϕ}{∂t} = \nu \frac{∂^2ϕ}{∂x^2}$$

Siendo esta última la ecuación del calor.

##**Burgers con diferencias finitas**

La ecuacion de Burgers es

$$u_t + u u_x - ν u_{xx} = 0$$

Con Euler centrado aproximo las derivadas equispaciales:

$$u_x = \frac{u_{j+1}^{n}-u_{j-1}^{n}}{2\Delta x}$$

$$u_{xx} = \frac{u_{j+1}^{n}-2u_{j}^{n}-u_{j-1}^{n}}{\Delta x ²}$$

y con Euler adelantado las temporales:

$$u_t = \frac{u_j^{n+1}-u_j^{n}}{\Delta t}$$

Juntando todo obtendria:

$$ u_j^{n+1} = Δt \biggm(-u_j^{n} \frac{u_{j+1}^{n}-u_{j-1}^{n}}{2\Delta x} + \nu \frac{u_{j+1}^{n}-2u_{j}^{n}+u_{j-1}^{n}}{\Delta x ²} \biggm) + u_j^{n} $$

```python
#Matrices de mis derivadas

def MatrizD1(N):
  M1 = np.zeros ( (N,N) )
  n = N-1
  M1[0,n] = -1
  M1[0,0] = 0
  M1[0,1] = 1
  for a in range(1,n):
    M1[a,a] = 0
    M1[a,a-1] = -1
    M1[a,a+1] = 1

  M1[n,n] = 0
  M1[n,0] = 1
  M1[n,n-1] = -1
  return M1


def MatrizD2 (N):
  M2 = np.zeros ( (N,N) )
  n = N-1
  M2[0,n] = 1
  M2[0,0] = -2
  M2[0,1] = 1
  for a in range(1,n):
    M2[a,a] = -2
    M2[a,a-1] = 1
    M2[a,a+1] = 1

  M2[n,n] = -2
  M2[n,0] = 1
  M2[n,n-1] = 1
  return M2
```

```python
def evol(u,dx,dt,nu):
  uxx = (MatrizD2(len(u))@u)/(dx**2)
  ux = (MatrizD1(len(u))@u)/(2*dx)
  return u + dt * (-1*u*ux + nu*uxx) #

def trayectoria(X0,dx,dt,nu,n):
  u = []
  for i in range(n):
    u.append(X0)
    Xs = evol (X0,dx,dt,nu)
    X0 = Xs
  return u
```

##**Burgers con métodos pseudoespectrales**

```python
import numpy as np

a = 1
nu = 0
N = 512
dt = .001
step = N

x = np.linspace(0,2*np.pi,N,endpoint=False) # Coordenada espacial en [0,2*pi)
t = np.arange(step)*dt                      # Tiempo
k = np.arange(0,N/2+1)

u = np.sin(x*a)
```

```python
def evol(u,k,nu,N,dt):
    # Evoluciona en el tiempo la PDE usando el método pseudoespectral y Runge-Kutta de segundo orden
    f  = np.fft.rfft(u)     #vaya al espacio de Fourier
    ft = f
    for ord in [2,1]:       # Bucle para el Runge-Kutta
        D   = dt/ord                 #D está siendo iterado entre D = {dt/2, dt/1}
        fux = compute_advection(f,k,N)      # Calculo la advección osea U*Ux
        fuxx = f*nu*((1j*k)**2)        #derivada de la advección en x, baja el argumento de la exponencial
        fev = (-1) * fux + fuxx     #evolución con euler
        f = ft + D * fev
        f[int(N/3):] = 0    # Dealiasing (eliminemos modos espúreos!)
    out = np.fft.irfft(f)  #Vuelva del espacio de Fourier
    return out

def compute_advection(f,k,N):
    # Calcula v*vx usando el método pseudoespectral y devuelve el resultado en el espacio Fourier
    fx  = 1j*k*f                      #derivamos
    v   =  np.fft.irfft(f)            # vuelva para poder multiplicar!
    vx  =  np.fft.irfft(fx)                       # eso de arriba.
    v   = v*vx
    out = np.fft.rfft(v)
    return out
```

```python
a = 1
nu = 0
N = 512
dt = 1e-3
t_f = 4
u = np.sin(x*a)


step = int(t_f / dt) #los puntos
x = np.linspace(0,2*np.pi,N,endpoint=False) # Coordenada espacial en [0,2*pi)
t = np.arange(step)*dt                      # Tiempo
k = np.arange(0,N/2+1)

out = np.zeros((N,step))
out[:,0] = u

for i in np.arange(step-1): # Evolución temporal
    out[:,i+1] = evol(out[:,i],k,nu,N,dt)


import matplotlib.pyplot as plt

plt.figure(figsize=(20, 8))
plt.suptitle('Analísis evolución bajo condición inicial senoidal para tf = 4', fontsize=16)

for i in np.arange(step-1):
  plt.plot(x, out[:,i], color ="g")
  plt.grid(linestyle='--')
  plt.grid(which = 'minor',linestyle=':', linewidth='0.1', color='black' )
plt.ylabel('u(x,t)')
plt.xlabel('x')
plt.show()
```

Definimos la clase MLP

```python
class MLP(torch.nn.Module):
    """
    Multilayer perceptron (MLP) // Perceptríon Multicapa .

    Esta clase define una red neuronal feedforward con múltiples capas ocultas
    lineales, funciones de activación tangente hiperbólica en  las capas ocultas
    y una salida lineal.

    Args:
        sizes (lista): Lista de enteros que especifica el número de neuronas en
        cada capa. El primer elemento debe coincidir con la dimensión de entrada
        y el último con la dimensión de salida.

    Atributos:
        capas (torch.nn.ModuleList): Lista que contiene las capas lineales del MLP.

    Métodos:
        forward(x): Realiza una pasada hacia adelante a través de la red MLP.

    Ejemplo:
        tamaños = [entrada_dim, oculta1_dim, oculta2_dim, salida_dim]
        mlp = MLP(tamaños)
        tensor_entrada = torch.tensor([...])
        salida = mlp(tensor_entrada)
    """
    def __init__(self,sizes):
        super().__init__()
        self.layers = torch.nn.ModuleList()
        for i in range(len(sizes)-1):
            self.layers.append(torch.nn.Linear(sizes[i],sizes[i+1]))
    def forward(self,x):
        h = x
        for hidden in self.layers[:-1]:
            h = torch.tanh(hidden(h))
        output = self.layers[-1]
        y = output(h)
        return y
```

```python
ptos = 200
x = np.linspace(0,2*np.pi,ptos,endpoint=False)
u = np.sin(x)
t = 4

#CI
t_data_init  =  torch.zeros(ptos).view(-1,1) #defina 100 puntos en 0   No olvidar .view(-1,1)
x_data_init  =  torch.linspace(0,2*np.pi,ptos).view(-1,1) #defina 100 puntos entre 0 y 2pi No olvidar .view(-1,1)
y_data_init  =  torch.tensor(u).view(-1,1) #defina la condicion inicial     No olvidar .view(-1,1)
input_init = torch.cat((x_data_init.flatten().view(-1,1), t_data_init.flatten().view(-1,1)), dim=1) #Junta las condiciones iniciales para que tengan la forma correcta
```

```python
#CC
t_data_cc    = torch.cat((torch.linspace(0,t,ptos), torch.linspace(0,t,ptos)), dim = 0).view(-1,1) #100 puntos temporales dos veces
x_data_cc    = 2*torch.pi* torch.cat((torch.zeros(ptos), torch.ones(ptos)), dim = 0).view(-1,1) #100 puntos en 0 y 100 en 1
y_data_cc    = torch.cat((torch.zeros(ptos), torch.zeros(ptos)), dim = 0 ).view(-1,1) #200 puntos que valen 0
input_cc = torch.cat((x_data_cc.flatten().view(-1,1), t_data_cc.flatten().view(-1,1)), dim=1) #Los juntamos para que pueda ser input la red
```

Definimos los puntos para evaluar la física. Estos deberán ser una grilla de 100x100 en el tiempo y el espacio.

```python
t_physics = torch.linspace(0,t,ptos) #100 puntos de 0 a 1
x_physics = torch.linspace(0,2*np.pi,ptos) #100 puntos de 0 a 2 pi
x_grid, t_grid = torch.meshgrid(x_physics, t_physics, indexing='ij')
x_grid = x_grid[:,:,None].requires_grad_(True) #Agregamos una dimensión al final para que pueda ser input de la red
t_grid = t_grid[:,:,None].requires_grad_(True) #Agregamos una dimensión al final para que pueda ser input de la red
input_physics = torch.cat((x_grid, t_grid), dim=-1)
```

```python
#Definición de la red y optimizador
BRGS = MLP([2,20,20,20,20,20,20,20,20,1])
BRGS.load_state_dict(torch.load("/content/drive/My Drive/Colab/Sistemas complejos/Guía 02/entrenamiento_g02.pth"))
optimizer = torch.optim.Adam(BRGS.parameters(),lr=1e-4)
```

Loop de entrenamiento

```python
BRGS.load_state_dict(torch.load("/content/drive/My Drive/Colab/Sistemas complejos/Guía 02/entrenamiento_g02.pth"))
iterations =  40000  #Numero de iteraciones
l =  1  #Peso de mi mod físico
nu = 1e-2

i = -1
for epoch in range(iterations):
    optimizer.zero_grad()
    #Condiciones iniciales
    yh = BRGS(input_init)     #Evalue las condiciones iniciales en la red
    loss1 = torch.mean((yh-y_data_init)**2)  #Defina este error como el MSE de la condición inicial
    #condiciones de contorno
    yh = BRGS(input_cc)    #Evalue las condiciones de contorno en la red
    loss2 = torch.mean((yh-y_data_cc)**2)   #Defina este error como el MSE de las condiciones de contorno
    yhp = BRGS(input_physics)     #Evalue las colocaciones de la física en la red
    dx  = torch.autograd.grad(yhp, x_grid, torch.ones_like(yhp), create_graph=True)[0]# computamos u_x
    dx2  = torch.autograd.grad(dx, x_grid, torch.ones_like(yhp), create_graph=True)[0]# computamos u_xx
    dt  = torch.autograd.grad(yhp, t_grid, torch.ones_like(yhp), create_graph=True)[0]# computamos u_t
    physics = dt + yhp*dx - nu*dx2 #Complete con la ecuación diferencial
    loss3 = l*torch.mean(physics**2)   #Defina este error como el MSE de la ecuación diferencial en las colocaciones pesado por l
    loss = loss1 + loss2 + loss3  #Sumamos todos los errores
    loss.backward()
    optimizer.step()
    i += 1
    if i%10 == 0:
      torch.save(BRGS.state_dict(),"/content/drive/My Drive/Colab/Sistemas complejos/Guía 02/entrenamiento_g02.pth")
      print("Guardado")
    with torch.autograd.no_grad():
    	print(epoch,'CI',float(loss1), 'CC', float(loss2), 'Física:',float(loss3),"Traning Loss:",float(loss.data))

torch.save(BRGS.state_dict(),"/content/drive/My Drive/Colab/Sistemas complejos/Guía 02/entrenamiento_g02.pth")
```

##**Ecuación de Kardar-Parisi-Zhang** (Burgers potencial)

La ecuación de Burgers en forma potencial puede hallarse considerando $\overline{u} = \overline{\nabla}\phi$:

\begin{equation*}
    \frac{\partial \phi}{\partial t} + \frac{1}{2}(\overline{\nabla}\phi)^2 - \nu \nabla^2 \phi.
\end{equation*}

A partir de esto, el modelo de Kardar-Parisi-Zhang (KPZ) se obtiene mediante un cambio de variables simple para $\phi(x,t)$ y agregando difusión y ruido al azar:

\begin{equation*}
    \frac{\partial h}{\partial t} + \frac{\lambda^2}{2}(\overline{\nabla}h)^2 - \nu \nabla^2 h + \eta,
\end{equation*}

donde $\eta$ representa un ruido Gaussiano.

Dada la versión lineal de KPZ (Burgers potencial) con forzado aleatorio

\begin{equation*}
    \frac{\partial \phi}{\partial t} +  \frac{1}{2}(\frac{\partial \phi}{\partial x})^2-\nu \frac{\partial^2 \phi}{\partial x^2}+\eta
\end{equation*}

donde $\eta(x,t)$ es ruido aleatorio con valor medio nulo, se pide:









###**Bloques para la integración temporal**


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def evol(u,k,nu,N,dt):
    # Evoluciona en el tiempo la PDE usando el método pseudoespectral y Runge-Kutta de segundo orden
    f   = np.fft.rfft(u)
    ft  = f
    eta = compute_random(N) # Calculamos el ruido para cada iteración completa de Runge-Kutta
    for ord in [2,1]:       # Bucle para el Runge-Kutta
        D   = dt/ord #vuelvo a usar un dt variable como en el 3
        fx  = 1j*k*f        # Derivada espacial de u en el espacio Fourier, baja el argumento de la exp
        fxx = -1*(k**2)*f #derivada segunda de u
        fnl = compute_vv(fx,N)
        f   = ft + D*(-eta+nu*fxx-0.5*fnl)               #integrador temporal
        f[int(N/3):] = 0    # Dealiasing
    out = np.fft.irfft(f)
    return out

def compute_vv(fx,N):
    # Calcula v^2 usando el método pseudoespectral y devuelve el resultado en el espacio Fourier
    ux   = np.fft.irfft(fx)
    out = np.fft.rfft(ux*ux)
    return out

def compute_random(N):
    # Genera ruido Gaussiano con media nula y varianza unitaria en el espacio Fourier
    phase  = 2*np.pi*np.random.rand(int(N/2+1))
    ampl   = np.random.randn(int(N/2+1))
    out    = ampl*np.exp(1j*phase)
    out[0] = 0.
    return out

def EspectKPZ(u,k,nu,Nx,step,dt):
  out = np.zeros((Nx,step)) #creo mi matriz que va a contener mi evolución
  out[:,0] = u
  for i in np.arange(step-1): # progreso temporalmente
      out[:,i+1] = evol(out[:,i],k,nu,Nx,dt)    #reemplazo los valores de mi matriz con ev, evolucionado = out[:,i]
  return out
```

```python
a = 0
A = 1e-2
nu = 1e-2


N = 512
dt = 1e-3
t_f = 2



step = int(t_f / dt) #los puntos
x = np.linspace(0,2*np.pi,N,endpoint=False) # Coordenada espacial en [0,2*pi)
t = np.arange(step)*dt                      # Tiempo
k = np.arange(0,N/2+1)

phi_0 = A*np.sin(a*x)

rto = EspectKPZ(phi_0,k,nu,N,step,dt)

plt.figure(figsize=(11,6))

for i in range(0,step,1000):
  plt.plot(x,rto[:,i],color = "orange")
plt.ylabel('u(x,t)')
plt.xlabel('x')
plt.grid()
plt.show()
```

Intento recuperar Burgers a partir de KpZ

Como KPZ es la forma potencial de Burgers, es decir que utilizamos $u = ∇𝚽 = \frac{∂𝚽}{∂x}$. Por lo tanto tengo que:

$$ u \cdot dx = 𝚽$$

Usando que en el ejercicio 3 utilicé un seno, propongo: $𝚽 = - cos(x)$.


```python
a = 1
A = -1
nu = 1e-2


N = 512
dt = 1e-3
t_f = 4



step = int(t_f / dt) #los puntos
x = np.linspace(0,2*np.pi,N,endpoint=False) # Coordenada espacial en [0,2*pi)
t = np.arange(step)*dt                      # Tiempo
k = np.arange(0,N/2+1)

phi_0 = A*np.cos(a*x)

rto = EspectKPZ(phi_0,k,nu,N,step,dt)


u = np.zeros((N,step))

for i in np.arange(step):
  f = np.fft.rfft(rto[:,i])
  fx = 1j*k*f
  f[int(N/3):] = 0
  out = np.fft.irfft(fx)
  u[:,i] = out
```
